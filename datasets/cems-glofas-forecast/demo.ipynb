```python
import earthkit
```

```python
data = earthkit.data.from_source(
    "cds",
    'cems-glofas-forecast',
    {
        'variable': 'river_discharge_in_the_last_24_hours',
        'system_version': 'operational',
        'hydrological_model': 'lisflood',
        'product_type': 'control_forecast',
        'year': '2023',
        'month': '01',
        'day': '01',
        'leadtime_hour': '120',
    },
)
```

```python
style = earthkit.maps.styles.Contour(
    colors="Blues",
    levels=[0.1, 1, 10, 50, 100, 500, 1000, 5000],
    extend="max",
)
```

```python
import cartopy.crs as ccrs
earthkit.maps.schema.figsize = (2, 2)
chart = earthkit.maps.Superplot(crs=ccrs.Robinson())

chart.contourf(data, style=style)

chart.coastlines(resolution="medium")
chart.gridlines(
    draw_labels=False,
    xlocs=range(-180, 180, 10),
    ylocs=range(-90, 90, 10),
    ylim=80,
    linestyle="-",
    alpha=0.5,
    linewidth=0.2,
)

chart.show()
```
